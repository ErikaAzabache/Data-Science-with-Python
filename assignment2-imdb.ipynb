```python
import pandas as pd
# , dtype={'startYear':'int', 'isAdult':'bool'}
basics = pd.read_csv('title.basics.tsv', sep='\t', index_col='tconst', usecols=['tconst', 'isAdult', 'startYear', 'genres'])
basics.head()
```

```python
episode = pd.read_csv('title.episode.tsv', sep='\t', index_col='tconst', usecols=['tconst', 'seasonNumber'])
episode.head()
```

```python
ratings = pd.read_csv('title.ratings.tsv', sep='\t', index_col='tconst', usecols=['tconst', 'averageRating'])
ratings.head()
```

```python
# merging basics, episodes, and ratings to obtain the initial dataframe
df = pd.merge(pd.merge(basics,episode, on='tconst'), ratings, on='tconst')
df
```

```python
# each record under genres contains a string with all the genres the title falls under.
# First we split the string into different columns per genre by title
mult_genre = df["genres"].str.split(",", n = -1, expand = True) 
mult_genre
```

```python
# We create a set of all possible genres in all multiple columns.
# This set will be used to create a column per unique genre 
# and a boolean to indicate if a title falls under that genre
len(mult_genre.columns)

all_genres = set()
for i in range(len(mult_genre.columns)):
    all_genres = all_genres | set(mult_genre[i].unique())
    
print(all_genres)
```

```python
# any missing or unspecified data is treated as the string "Empty"
# a column per unique genre is created (boolean)
for genre in all_genres:
    g = str(genre)
    if g in ['nan', '\\N', 'None']: g = "Empty"
    print(genre)
    df["gr_" + g] = df.genres.str.contains(r'%s' % g)
```

```python
# This only to see how seasonNumber looks like
# all_seasons = set(df['seasonNumber'].unique())
#len(all_seasons)

# convert columns to integers, if value is '\\N', then assume 0
df['seasonNumber'][ df['seasonNumber']=='\\N' ] = 0
df['seasonNumber'].astype(int)

df['startYear'][ df['startYear']=='\\N' ] = 0
df['startYear'].astype(int)
```

```python
# remove genres string colummn in lieu of booleans by genre
N_SAMPLES = 100
df_n = df.drop('genres', axis=1)[:N_SAMPLES]
df_n
```
